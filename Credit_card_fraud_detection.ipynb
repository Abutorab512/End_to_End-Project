But we don't want our model to assume, we want our model to detect patterns that give signs of fraud!

```python
from sklearn.preprocessing import RobustScaler

rs = RobustScaler()
df['Time'] = rs.fit_transform(df['Time'].values.reshape(-1,1))
df['Amount'] = rs.fit_transform(df['Amount'].values.reshape(-1,1))
```

```python
# Since our classes are highly skewed we should make them equivalent in order to have a normal distribution of the classes.

# Lets shuffle the data before creating the subsamples

df = df.sample(frac=1)

# amount of fraud classes 492 rows.
fraud_df = df.loc[df['Class'] == 1]
non_fraud_df = df.loc[df['Class'] == 0][:492]

normal_distributed_df = pd.concat([fraud_df, non_fraud_df])

# Shuffle dataframe rows
new_df = normal_distributed_df.sample(frac=1, random_state=42)

new_df.head()
```

```python
print('Distribution of the Classes in the subsample dataset')
print(new_df['Class'].value_counts()/len(new_df))



sns.countplot('Class', data=new_df, palette=colors)
plt.title('Equally Distributed Classes', fontsize=14)
plt.show()
```

```python
x = df.drop(['Class'],axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=142)
```

```python
print(x_train.shape,x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,model.predict(x_test))
```

```python
accuracy_score(y_train,model.predict(x_train))
```

```python
confusion_matrix(y_test,model.predict(x_test))
```

```python
confusion_matrix(y_train,model.predict(x_train))
```
