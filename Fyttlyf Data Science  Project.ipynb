#### we need to convert all categorical data into numerical form by using label encoding

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df['MobileWeb_or_Web'] = lb.fit_transform(df['MobileWeb_or_Web'])
df['Type_of_Customers?'] = lb.fit_transform(df['Type_of_Customers?'])
df['Where_Are_They_comming_from?'] = lb.fit_transform(df['Where_Are_They_comming_from?'])
df['Which_Place_in_India?'] = lb.fit_transform(df['Which_Place_in_India?'])
```

```python
df.drop(['div of two columns'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.describe().round(1)
```

```python
x = df.drop(['How_many_Landed_on_the_our_Page_and_clicked_on_a_button_and_started_filling_the_Form_and_Completed_and_submited_the_form?'],axis=1)
```

```python
y = df['How_many_Landed_on_the_our_Page_and_clicked_on_a_button_and_started_filling_the_Form_and_Completed_and_submited_the_form?']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20, random_state=125)
```

```python
(x_train.shape),(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,model.predict(x_test)).round(2)
```

```python
mean_absolute_percentage_error(y_train,model.predict(x_train)).round(2)
```
